```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = '/content/drive/MyDrive/Coding Dojo/Week 7/Diamonds Prices2022.csv'

df = pd.read_csv(path)
df.head()
```

There are 53,940 diamonds in the dataset with 10 features (carat, cut, color, clarity, depth, table, price, x, y, and z). Most variables are numeric in nature, but the variables cut, color, and clarity are ordered factor variables with the following levels.
About the currency for the price column: it is Price ($)
And About the columns x,y, and z they are diamond measurements as (( x: length in mm, y: width in mm,z: depth in mm ))

```python
# The unnamed column is unnecessary so I decided to drop it.

diamond_df = df.drop(columns=['Unnamed: 0'])
diamond_df.head()
```

```python
diamond_df.info()
```

```python
#Checking to see if there any difficult values
diamond_df.describe()
```

```python
#The data has no missing values. 
diamond_df.isna().sum()
```

```python
# Assigning the y (target) and X (features)
y = diamond_df['price']
X = diamond_df.drop(columns= 'price')
```

```python
X.head()
```

```python
plt.figure(figsize=(14,8))
plt.bar(y, diamond_df['carat'])
plt.xlabel('Price in USD', fontsize= 15)
plt.ylabel('Carats', fontsize= 15)
plt.title('Price of Diamond Per Carats', fontsize=18)
plt.xlim(200,19000)
plt.grid();
```

The more the carats the more expensive the diamond is. However there are some outliers where carat was 5 but the price averaged about 13500 USD. There was also when it was 2 carat and it was about 8000 USD

```python
plt.figure(figsize=(10,8))
plt.plot(y,diamond_df['x'], label='Length in mm')
plt.plot(y,diamond_df['y'], label ='Width in mm')
plt.plot(y,diamond_df['z'], label= 'depth in mm')
plt.xlabel('Price', fontsize= 15)
plt.ylabel('Count in mm', fontsize = 15)
plt.legend()

plt.title('Price of Diamond According to Length, Width & Depth', fontsize= 18)
plt.grid();
```

The more the length, width and depth the more expensive the diamond will be. However there some are outliers for the price of the diamond where width was 58mm (the max width) but it was about 12000$. 

Also there is an outlier where max depth and width of 31mm went for about 2000$.
